## Approach

### Workflow
1. Spreading:
    - Input $x \in R^N$ is being spread to M dimensions using a spreading matrix $S \in R^{NxM}$
2. RBF Activation
    - Pass through $M$ gaussian functions with mean $\mu_i$ and var $\sigma_i$
3. Output


### Initialization
$\mu$ and $\sigma$ are initialized in a "Spektral" manner to have multiple gaussians covering the same space with different variances.



```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_gaussian_parameters(levels, max_range):
    """
    Calculate the means and variances for Gaussians based on the specified levels and range.

    :param levels: Number of levels of Gaussians
    :param max_range: Maximum range for the x-values
    :return: Lists of means and variances for each level
    """
    means = []
    variances = []

    # Starting variance
    initial_variance = max_range*2

    for level in range(1, levels + 1):
        # Calculate the variance for this level
        variance = initial_variance / level

        # Calculate the means for this level
        if level == 1:
            # For the first level, the mean is 0
            means_level = [0]
        else:
            step = max_range / (level - 1)
            means_level = [ (step / 2 + step * i  - max_range) for i in range(level-1)]

        # Extend the means and variances lists
        means.extend(means_level)
        variances.extend([variance] * len(means_level))

    return means, variances

# Calculate means and variances for 3 levels within a range of -10 to 10
means, variances = calculate_gaussian_parameters(3, 10)

# Generate x values for plotting
x = np.linspace(-15, 15, 1000)

# Plot each Gaussian
plt.figure(figsize=(10, 6))
for mean, var in zip(means, variances):
    y = (1 / np.sqrt(2 * np.pi * var)) * np.exp(-0.5 * ((x - mean) ** 2) / var)
    plt.plot(x, y, label=f'Mean = {mean}, Var = {var}')

# Adding labels and title
plt.title('Calculated Gaussians with Different Means and Variances')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Show plot
plt.show()

```

```python
## plotting
import numpy as np
import plotly.graph_objects as go

data = experiment_data_loader.experiment_data['000008']['repetition_data'][1]


means = []
for mean_idx in range(16):
    means.append(data[f"mean{mean_idx}"])

    
data = np.stack(means, axis=1)
print(data.shape)
#data = np.expand_dims(data, axis=1)

# Preparing data for plotting
epochs = np.arange(1, data.shape[0])
traces = []
for i in range(data.shape[1]):
    traces.append(go.Scatter(x=epochs, y=data[:, i], mode='lines', name=f'Value {i+1}'))

# Plotting with Plotly
layout = go.Layout(title='Change of 16 Values over 500 Steps',
                   xaxis=dict(title='Epoch'),
                   yaxis=dict(title='Value'),
                   template='plotly_dark')

fig = go.Figure(data=traces, layout=layout)
fig.show()

```

### Issues

- Change during Backpropagation is dependent on the distance of the input value to the mean:

$$ \frac{\partial o_j}{\partial \mu_i} = \frac{\mu_i}{\sigma^2} e^{-\frac{1}{2} (\frac{||x-\mu_i||}{\sigma})^2} $$

=> We might get "dead areas" when the model is trained on a dataset and new datapoints arise outside of the trained dataset
